```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt

DataIn = np.loadtxt('sunspots.txt')
mes = DataIn[:,0]
manchas = DataIn[:,1]

mes = 1749 + (mes/12)

plt.scatter(mes, manchas, label="Sunspots")
plt.xlabel("Ao")
plt.ylabel("Manchas solares")
plt.xticks(mes[12::25*12]);
plt.legend(loc='best')

mes[-2:]
```

```python
pM = manchas

for i in range (1,len(pM)-1):
    pM[i] = (manchas[i-1] + manchas[i] + manchas[i+1])/3

plt.scatter(mes, manchas, label="Sunspots")
plt.xlabel("Ao")
plt.ylabel("Manchas solares")
plt.xticks(mes[12::25*12]);
plt.plot(mes,pM, color="red",linewidth=1.0, label="Promedio Mvil")
plt.legend(loc='best')
```
