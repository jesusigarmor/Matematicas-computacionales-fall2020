```r
flights
```

```r
#Sacar el mejor día de la semana para volar
library(lubridate)
flights %>%
    mutate(week_day=wday(mdy(paste0(month,'-',day,'-',year)))) %>%
    group_by(week_day) %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
```

```r
#¿Hay alguna aerolínea que tenga algún problema a nivel mes?

flights %>%
    group_by(carrier,month) %>%
    summarise(
        delay=mean(dep_delay,na.rm=TRUE )
    ) %>%
    arrange(desc(delay))
```

```r
#¿Hay algún avión problemático?

flights %>%
    group_by(tailnum)  %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
    
```

```r
#¿Hay algún destino/origen problemático?

flights %>%
    group_by(origin)  %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
```

```r
flights %>%
    group_by(dest)  %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
```

```r
#¿Hay alguna correlación entre distancia y retrasos?
flights %>%
    group_by(distance)  %>%
    summarise(delay=mean(dep_delay,na.rm=TRUE)) %>%
    arrange(desc(delay))
```
